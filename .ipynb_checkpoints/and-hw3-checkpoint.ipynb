```python
def extract_single_place(url):
    result = requests.get(url)
    soup = bs(result.text)
    title = soup.find_all('h1', {'class':'DDPage__header-title'})[0].contents[0]
    short_description = soup.find_all('h3', {'class': 'DDPage__header-dek'})[0].contents[0]
    
    location_tag = soup.find_all('div', {'class':'DDPage__header-place-location'})[0].contents[0]
    if type(location_tag) == bs4.element.NavigableString:
        location = location_tag.text
    else:
        location = location_tag.contents[0].text
    
    long_description = [x.text for x in soup.find_all('div', {'id': 'place-body'})]
    
    return {'title': title, 
            'location': location, 
            'short_description':short_description,
            'long_description':long_description}
```

```python
list_page = requests.get(url)
list_soup = bs(list_page.text)
list_places = [x.get('href') for x in list_soup.find_all('a', {'class':'content-card-place'})]
data = {}
for place in tqdm(list_places):
    if not place in data:
        extract = extract_single_place('https://www.atlasobscura.com' + place)
        data[place] = extract
```

```python
import os
# psi = [psi[i : i + 4] for i in range(0, len(psi), 3)][:-1]
        
def url_to_text_(urls):
    i = urls[0]
    url = urls[1].split('\n')[0]
    try:
        os.makedirs('htmls_page_{}'.format(i+1))
    except:
        pass
    el = url
    el_name = el.split('.com/places/')[1]
    result = requests.get(el)
    with open('./htmls_page_{}/{}'.format(i+1, el_name), 'w') as wf:
        wf.write(result.text)
    wf.close()
         
        
if __name__ == '__main__':
    with open ('txt_file_urls.txt', 'r') as urls:
        l = []
        for line in urls:
            l.append(line)
        grouped_l = [l[i:i+18] for i in range(0,len(l), 18)]
        grouped_ln = [[(i, el) for el in group]for i, group in enumerate(grouped_l)]

        grouped_ln_flat = []
        for group in grouped_ln:
            for el in group:
                grouped_ln_flat.append(el)
         
        for el in grouped_ln_flat[(170*18):(170*18)+18]:
            url_to_text_(el)
```
